```python
# Display data frame for csv file "worldcities1.csv"

world_cities = pd.read_csv("data/worldcities1.csv")
world_cities.head()
#world_cities
```

```python
# Do a group by based on  column name "country"

world_cities_groupby = world_cities.groupby("country")["country","population"]
# To display maximum rows
pd.set_option('display.max_rows', None)
#world_cities_groupby.head()
```

```python
# Calculating total population for every country available in the dataset

population_by_country = pd.DataFrame(world_cities_groupby.sum()["population"]).reset_index(["country"])
#population_by_country.head()
population_by_country
```

```python
# Calculating  number of countries

world_cities_count = population_by_country.count()
world_cities_count
```

```python
# Display data frame for the csv file "happy.csv"

happy = pd.read_csv("data/happy.csv")
happy.head()
#happy
```

```python
# Sort by column name "country" and reset index

happy_sort_reindexed = happy.sort_values("country", ascending=True).reset_index(drop=True)
pd.set_option('display.max_rows', None)
happy_sort_reindexed
```

```python
# Calculating the number of countries

happy_count = happy_sort_reindexed.count()
happy_count
```

```python
# Merging the two dataframes "population_by_country" and "happy_sort_reindexed"
# Doing right join because considering only the details of the countries which are available in the "happy.csv" file

merge_worldcities_happy = pd.merge(population_by_country, happy_sort_reindexed, on="country", how="right")
merge_worldcities_happy
```

```python
# Calculating the number of countries

merge_worldcities_happy_count = merge_worldcities_happy.count()
merge_worldcities_happy_count
```
