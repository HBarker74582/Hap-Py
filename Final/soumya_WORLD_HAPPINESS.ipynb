##  Analyzing Average Happiness Score By Region(based on datasets 2015,2016,2017,2018)

Obtain a bar plot of the Average Happiness Score by Region from year 2015 to 2018. Bar chart have the Average happiness score of each region listed vertically along the y-axis and the x-axis have region.In this plot we are analyzing 4 years happiness score by region.

```python

# Concatinating happiness dataframes "whr_2015" , "whr_2016" , "whr_2017" , "whr_2018"
whr_combined_four_years = pd.concat([whr_2015, whr_2016, whr_2017_region, whr_2018_region], ignore_index=True, sort=False)

# Taking columns "Year" , "Region" , and "Happiness Score" from the concatinated dataframe
whr_combined_four_years_happines_by_region = whr_combined_four_years[['Year', 'Region', 'Happiness Score']]

# Do a groupby for "Year" and "Region"
# Taking mean or average of the "Happiness Score" column
# Adding suffix "_Avg" to the "Happiness Score" column
# Reset index
whr_combined_four_years_happines_by_region_groupby = whr_combined_four_years_happines_by_region.groupby(['Year','Region']).mean().add_suffix('_Avg').reset_index()
whr_combined_four_years_happines_by_region_groupby.head()
```

```python
# Plotting Happiness Score by region (2015 to 2018) 
# USING PANDAS PIVOT TABLE FOR PLOTTING MULTIPLE YEAR BAR GRAPH

#The simplest pivot table must have a dataframe and an index
# Here we have single index "Region"
# We can have multiple indexes as well
# Most of the pivot table arguments can take multiple values via a list
# Here we have taken "Year" as column
# plotting for unique rows (2015,2016,2017,2018) from column "Year"
# Here we are taking Average Happiness Score on y axis

Happiness_by_region_2015_to_2018 = pd.pivot_table(whr_combined_four_years_happines_by_region_groupby,
                                                  index = 'Region', columns = 'Year', values = 'Happiness Score_Avg')
Happiness_by_region_2015_to_2018.plot(kind = 'bar', figsize=(15,8), fontsize = 15, rot=90)

#set title
plt.title('Happiness Score by Region (Years 2015 to 2018)', fontsize = 15)
#set legend
plt.legend(bbox_to_anchor=(0, 1), loc='right', ncol=1, fontsize = 12)
#set yaxis label
plt.ylabel('Average Happiness Score', fontsize = 15)
# set xaxis label
plt.xlabel('Region', fontsize = 15)
# To save figure      
plt.savefig("soumya_Output/Happiness_by_region_2015_to_2018.png", bbox_inches="tight") 
# show figure
plt.show()
```

## Analyzing Happiness score of Top fifty countries(from 2018 data) by region(2015-2018)


Obtain a bar plot of the Happiness Score of Top fifty countries (from 2018 data) Vs Region for period of 2015 to 2018. Bar chart have the Country count(number of countries) of each region listed vertically along the y-axis and the x-axis have region.In this plot we are analyzing 4 years happiness score by region.

```python
# Top fifty ranks condition statement
Top_fifty_condition = whr_combined_four_years['Happiness Rank'] < 51

# Applying top fifty where condition on combined whr data
whr_combined_four_years_top_50 = whr_combined_four_years[Top_fifty_condition]
whr_combined_four_years_top_50

# Taking columns "Year" , "Region" , and "Happiness Score" from the top fifty countries
whr_combined_four_years_top_50_by_region = whr_combined_four_years_top_50[['Year', 'Region', 'Country']]

# Do a groupby for "Year" and "Region"
# Taking count of the "Country" column
# Adding suffix "_Count" to the "Country" column
# Reset index
whr_combined_four_years_top_50_by_region_groupby = whr_combined_four_years_top_50_by_region.groupby(['Year','Region']).count().add_suffix('_Count').reset_index()
whr_combined_four_years_top_50_by_region_groupby.head()
```
