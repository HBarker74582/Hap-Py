```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

mich_path1 = 'data/one-star-michelin-restaurants.csv'
mich_path2 = 'data/two-stars-michelin-restaurants.csv'
mich_path3 = 'data/three-stars-michelin-restaurants.csv'
happ_path = 'data/happy.csv'
```

```python
mich1_clean = pd.read_csv(mich_path1)
del mich1_clean['price']
del mich1_clean['url']
del mich1_clean['longitude']
del mich1_clean['latitude']
del mich1_clean['zipCode']
# newyork, chicago and cali need to be combined into unitedstates
mich1_clean['country'] = mich1_clean['country'].replace(
    {'California' : 'United States', 'Chicago' : 'United States', 'New York City' : 'United States', 
     'Washington DC' : 'United States', 'Hong Kong' : 'China', 'Macau' : 'China', 'Taipei' : 'Taiwan', 
     'Rio de Janeiro' : 'Brazil', 'Sao Paulo' : 'Brazil' })
```

```python
mich1_count = mich1_clean['country'].value_counts()
mich1_avg = mich1_count.mean()
finland_df1 = mich1_clean.loc[mich1_clean['country'] == 'Finland']
fin1_count = finland_df1['name'].count().round()
 
mich1_avg
```

```python
mich2_clean = pd.read_csv(mich_path2)
del mich2_clean['price']
del mich2_clean['url']
del mich2_clean['longitude']
del mich2_clean['latitude']
del mich2_clean['zipCode']

mich2_clean['country'] = mich2_clean['country'].replace(
    {'California' : 'United States', 'Chicago' : 'United States', 'New York City' : 'United States', 
     "Washington DC" : "United States", "Hong Kong" : "China", "Macau" : "China", "Taipei" : " Taiwan", 
     "Rio de Janeiro" : "Brazil", "Sao Paulo" : "Brazil" })
mich2_count = mich2_clean['country'].value_counts()
```

```python
mich2_avg = mich2_count.mean()
finland_df2 = mich2_clean.loc[mich2_clean['country'] == 'Finland']
fin1_count2 = finland_df2['name'].count()
mich2_count
```

```python
mich3_clean = pd.read_csv(mich_path3)
del mich3_clean["price"]
del mich3_clean['url']
del mich3_clean['longitude']
del mich3_clean['latitude']
del mich3_clean["zipCode"]

mich3_clean['country'] = mich3_clean['country'].replace(
    {'California' : 'United States', 'Chicago' : 'United States', 'New York City' : 'United States', 
     'Washington DC' : 'United States', 'Hong Kong' : 'China', 'Macau' : 'China', 'Taipei' : 'Taiwan', 
     'Rio de Janeiro' : 'Brazil', 'Sao Paulo' : 'Brazil'})

mich3_count = mich3_clean['country'].value_counts()
```

```python
mich3_avg = mich3_count.mean()
finland_df3 = mich3_clean.loc[mich3_clean['country'] == 'Finland']
fin1_count3 = finland_df3['name'].count()
mich3_count.head()
```

```python
happy_clean = pd.read_csv(happ_path)
happy_cleaner = happy_clean[['country', 'Ladder']]
# happy_cleaning=happy_cleaner.rename(columns={'country': 'region'})
```

```python
mich1_by_name = mich1_clean.groupby('country').count()['name']
mich2_by_name = mich2_clean.groupby('country').count()['name']
mich3_by_name = mich3_clean.groupby('country').count()['name']
m1 = pd.merge(mich1_by_name, mich2_by_name, how='left', on=['country'])
m2 = pd.merge(m1, mich3_by_name, how='left', on=['country'])
m2 = m2.fillna(0)
m2
```

```python
m3= m2.sum(axis=1)
m4= pd.DataFrame(m3).reset_index()
m4=m4.rename(columns={ 0: 'restaurant count'})
```

```python
m5 = pd.merge(happy_cleaner, m4, how='left', on='country')
# m5.rename(columns={'0': 'restaurant count', 'region' : 'country'})
m5 = m5.fillna(0)
m5
```

```python
m5.count()
```

```python
 avg_rest_count = m5['restaurant count'].mean()
```

```python
top_20 = m5.truncate(before=0, after=20)
low_20 = m5.truncate(before=136, after=155)
low_20
```

```python
top_20_avg=top_20['restaurant count'].mean()
low_20_avg=low_20['restaurant count'].mean()
```

```python
x_axis = np.arange(len(top_20))
tick_locations = [value+0.4 for value in x_axis]
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(40,5))
plt.bar(x_axis, top_20["restaurant count"], color='orange', alpha=1, align="edge")
plt.xticks(tick_locations, top_20["country"], rotation="horizontal")
title=("Top 20 Restaurant Count")
plt.xlabel("Countries")
plt.ylabel('Restaurant Count')
plt.savefig("top_20_bar.png")
```

```python
x_axis = np.arange(len(low_20))
tick_locations = [value+0.4 for value in x_axis]
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(40,5))
plt.bar(x_axis, low_20["restaurant count"], color='green', alpha=0.5, align="edge")
plt.xticks(tick_locations, low_20["country"], rotation="horizontal")
plt.xlabel("Countries")
plt.ylabel('Restaurant Count')
plt.savefig("low_20_bar.png")
```
